```python
#1,1
class area:
    def __init__(self,a,b,c):
        self.a = a
        self.b = b
        self.c = c
    def form(self):
        s = (self.a + self.b + self.c)/2 
        return (s*(s-self.a)*(s-self.b)*(s-self.c)) ** 0.5
```

```python
a = area(5,2,1)
```

```python
a.form()
```

```python
#1.2
def filter_long_words(word,n):
    d = []
    for i in word:
        if len(i) > n:
            d.append(i)
    return d
```

```python
filter_long_words(['asd','zx','mnbvc','hjkl'],3)
```
